```python
import numpy as np
import time
for freq in np.arange(100,700,100):
    fstr = 'FREQ ' + "%s" % freq  + 'MHZ'
    olvl = 'OLVL ' + "%s" % (freq/10) + 'mV'
    print(fstr,' ',olvl)
    time.sleep(0.5)
    siggen.write(olvl)
    siggen.write(fstr)
    time.sleep(0.5)
#    siggen.write('LVL ON')
#    time.sleep(1)
#    siggen.write('LVL OFF')
```
