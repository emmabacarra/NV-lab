### PyVisa code for Anritsu MG3681 connection test

```python
import pyvisa
rm = pyvisa.ResourceManager()
print(rm)
rm.list_resources()
```

```python
siggen.read_termination = '\n'
siggen.write_termination = '\n'
siggen.baud_rate = 9600
siggen.delay = 0.25
```

```python
siggen = rm.open_resource('ASRL5::INSTR', baud_rate=9600, data_bits=8, 
                          read_termination = '\r\n', write_termination = '\n')
```

```python
siggen.query('FREQ?')
```

```python
siggen.write('FREQ 328.22 MHz')
```

```python
siggen.write('OLV')
```

```python
test = 50
siggen.write(f'OLVL {test}mV')
siggen.query('OLVL?')
```

```python
siggen.write('OLDBM')
siggen.query('OLVL?')
```

```python
siggen.write('LVL OFF')
```

```python
siggen.write('LVL ON')
```

```python
import numpy as np
import time
for freq in np.arange(100,700,1):
    fstr = 'FREQ ' + "%s" % freq  + 'MHZ'
    print(fstr)
    time.sleep(0.5)
    siggen.write(fstr)
    time.sleep(0.5)
#    siggen.write('LVL ON')
#    time.sleep(1)
#    siggen.write('LVL OFF')
```

```python
import numpy as np
import time
for freq in np.arange(100,700,100):
    fstr = 'FREQ ' + "%s" % freq  + 'MHZ'
    olvl = 'OLVL ' + "%s" % (freq/10) + 'mV'
    print(fstr,' ',olvl)
    time.sleep(0.5)
    siggen.write(olvl)
    siggen.write(fstr)
    time.sleep(0.5)
#    siggen.write('LVL ON')
#    time.sleep(1)
#    siggen.write('LVL OFF')
```

```python
siggen.query("FREQ?")
```

```python
siggen.query("OLVL?")
```
