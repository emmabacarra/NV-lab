# Test lockin detection of PL vs laser pulse length

Import Statements

```python
#Logging
import logging
log = logging.getLogger(__name__)
log.addHandler(logging.NullHandler())

#General Python stuff
import sys
import tempfile
import random
from time import sleep
import numpy as np

#Pymeasure 
from pymeasure.log import console_log
from pymeasure.display.Qt import QtGui
from pymeasure.display.windows import ManagedWindow
from pymeasure.experiment import Procedure, Results
from pymeasure.experiment import IntegerParameter, FloatParameter, Parameter

#Pulse blaster and signal generator 
from spinapi import *
import pyvisa
from Our_Functions import *
```

This is our procedure class. In this class, we can specify our inputs using pymeasure methods. We then create an execute method which is where the "meat" of our code is. This is where we turn our devices on and off, and record our data. We also write a pulsing function which has the specific pulse sequence for this experiment. We create this extra function just so that our execute method has fewer lines and is thus easier to read. 

```python
class Exp1(Procedure):

    TlaserMin = FloatParameter('TlaserMin', units = 'ns', default = 1*us)
    TlaserMax = FloatParameter('TlaserMax', units = 'ns', default = 200*us)
    MaxTdelay = FloatParameter('MaxTdelay', units = 'ns', default = 1*ms)
    Tref = FloatParameter('Tref', units = 'ns', default = 15*ms)
    num_delay_times = IntegerParameter('Number of unique Tdelays', default = 20)
    num_data_points = IntegerParameter('Number of points per delay time', default =1)
    sleep_time = FloatParameter('Measurement Duration', units = 's', default = 12)
    
    DATA_COLUMNS = ['Tlaser', 'Lock-in Amplitude']

    def execute(self):
        log.info("Starting the loop of %d iterations" % self.num_delay_times)
        
        Tlaser_arr = np.geomspace(self.TlaserMin, self.TlaserMax, self.num_delay_times)
        #Construct our pulse blaster object
        pb = PulseBlaster(0,500)
        pb.get_started()
        
        rm = pyvisa.ResourceManager()
        rm.list_resources()
        
        lockin = rm.open_resource('ASRL4::INSTR', baud_rate=9600, data_bits=8, 
                                  read_termination = '\r', write_termination = '\r\n')

        for Tlaser in range(0,self.num_delay_times):
            
            #Load the pulse program in
            pulse1(self.Tref, Tlaser_arr[Tlaser])
#            pulse1a(self.Tref, self.Tlaser, Tdelays_arr[Tdelays], pb)
                    
            #Give the code to load the pulse program
            sleep(.2)
            
            #Trigger the pulse blaster
            pb.trigger_blaster()
                            
            #Sleep to allow the lock-in to reach a steady value
            sleep(self.sleep_time)

            #Take lock-in measurement (measuring the R value)
            R = float(lockin.query('OUTP? 3'))
        
            #When we actually get data from our lock-in, this is where our data will be stored as a dictionarry
            data = {
#                    'Tdelay': Tdelays_arr[Tdelays] + self.MaxTdelay/(self.num_iters*self.num_data_points)*i,
                'Tlaser': Tlaser_arr[Tlaser],
                'Lock-in Amplitude': R*1000
            }
            #The data for this iteration is emitted to our results class that we will later construct and execute
            self.emit('results', data)
        
            #This is used so that if we want to stop the experiment early from the GUI, we can do so
            if self.should_stop():
                log.warning("Caught the stop flag in the procedure")
                break

            sleep(.5)
        
            pb_reset()
            
        #After we pulse, we stop and close the pulse blaster
        pb.stop_trigger()
        
def pulse1(Tref, Tlaser):
    
    #Initialize which bits are turned on and off 
    a = 0b000000000000000000110011
    b = 0b000000000000000000010001
    c = 0b000000000000000000100010
    d = 0b000000000000000000000000
    e = 0b000000000000000000100010
    f = 0b000000000000000000000000
    end = 0b000000000000000000000000

    #Program the pulse program ---------------------------------------------------------------
    pb_start_programming(PULSE_PROGRAM)
    
    #First initialization pulse
    start = pb_inst_pbonly(a, Inst.CONTINUE, 0, Tlaser)

    #Wait for half of the lock-in cycle
    pb_inst_pbonly(b, Inst.CONTINUE, 0, Tref - Tlaser)

    #Second initialization
    # pb_inst_pbonly(c, Inst.CONTINUE, 0, Tlaser)

    #Delay time
    # pb_inst_pbonly(d, Inst.CONTINUE, 0, Tdelay)

    #Readout pulse
    # pb_inst_pbonly(c, Inst.CONTINUE, 0, Tlaser)

    #Wait again and go back
    # pb_inst_pbonly(f, Inst.BRANCH, start, Tref - 2*Tlaser - Tdelay)
    pb_inst_pbonly(f, Inst.BRANCH, start, Tref)

    pb_stop_programming()
    #-----------------------------------------------------------------------------------------

    
def pulse1a(Tref, Tlaser, Tdelay, pb):
    
    start = pb.start_programming()
    
    for i in range(10):
        
        pb.command([devices.lockin, devices.laser], Tlaser)

        pb.command([devices.lockin], Tref/10 - Tlaser)

    for i in range(10):
        pb.command([devices.laser], Tlaser)

        pb.command([devices.nothing], Tdelay)

        pb.command([devices.laser], Tlaser)

        pb.command([devices.nothing], Tref/10 - 2*Tlaser - Tdelay)
    
    pb.stop_programming(start)
    
    
```

This class creates a GUI from which we can enter our parameters and run the experiment. The data that is currently being plotted isn't real data, it's just random numbers

```python
class MainWindow(ManagedWindow):

    def __init__(self):
        super().__init__(
            
            #Specify our procedure class
            procedure_class = Exp1,
            
            #Decide which inputs and displays we want to use
            inputs=['Tref', 'MaxTdelay', 'num_delay_times', 'num_data_points', 'sleep_time'],
            displays=['Tref', 'MaxTdelay', 'num_delay_times', 'num_data_points', 'sleep_time'],
            
            #Configure our plot
            x_axis='Tlaser',
            y_axis='Lock-in Amplitude'
        )
        self.setWindowTitle('Experiment 1: Laser Pulses')

    def queue(self):
        filename = tempfile.mktemp()

        procedure = self.make_procedure()
        results = Results(procedure, filename)
        experiment = self.new_experiment(results)

        self.manager.queue(experiment)

#If this Jupyter Notebook is the notebook that contains the source code for our procedure (i.e. we didn't import it), then run the GUI
if __name__ == "__main__":
    app = QtGui.QApplication(sys.argv)
    window = MainWindow()
    window.show()
    sys.exit(app.exec_())
```
